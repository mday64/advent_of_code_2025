At this point, we have Row Echelon Form.  Is it possible to start seeking solutions at this point?

The bottom row has one free variable, which should make it easy to generate possible solutions (for that row).  Then try substituting into higher rows to solve for other variables.  But what if we encounter other free variables?

```python
r[5]-=r[6]
r
```

Because of the 3 in row r[5], I'm going to assume x[7] = 1 and x[6] = 14.  But this will cause a problem later, making x[1] = -10.  This will require x[7] to be larger.

```python
r = numpy.vstack([r, numpy.array([0., 0., 0., 0., 0., 0., 0., 1., 13.])])
r
```

```python
r[6] += r[7]
r[5] -= r[7]
r[3] += 2*r[7]
r[1] -= r[7]
r[0] -= r[7]
r
```

```python
r[5] /= 3.
r
```

```python
r[5] -= r[7]
r
```

```python
r[0] -= r[5]
r[1] -= r[5]
r[2] += r[5]
r[3] += 2 * r[5]
r
```
